```python
# Plotting multiple bar charts at once for categorical variables
# Since there is no default function which can plot bar charts for multiple columns at once
# we are defining our own function for the same

def PlotBarCharts(inpData, colsToPlot):
    %matplotlib inline
    
    import matplotlib.pyplot as plt
    
    # Generating multiple subplots
    fig, subPlot=plt.subplots(nrows=1, ncols=len(colsToPlot), figsize=(20,5))
    fig.suptitle('Bar charts of: '+ str(colsToPlot))

    for colName, plotNumber in zip(colsToPlot, range(len(colsToPlot))):
        inpData.groupby(colName).size().plot(kind='bar',ax=subPlot[plotNumber])
```

```python
# Calling the function for 5 columns
PlotBarCharts(inpData=data, 
              colsToPlot=['checkingAccount', 'creditHistory', 'purpose','savingsAccount','presentEmployment'])
```

```python
# calling the functions of 5 columns
PlotBarCharts(inpData=data,
             colsToPlot=['installmentRate','personalStatus','otherDebtors','presentResidence','property'])
```

```python
SelectedColumns=['checkingAccount','creditHistory','purpose','savingsAccount','presentEmployment',
 'personalStatus','otherDebtors','property','otherInstallments','housing','foreignWorker',
 'age', 'creditAmount', 'durationMonth']

# Selecting final columns
DataForML=data[SelectedColumns]
DataForML.head()
```

```python
# Saving this final data for reference during deployment
DataForML.to_pickle('DataForML.pk')
```

# Converting ordinal values to numeric values

```python
# Treating the ordinal variable first 
DataForML['presentEmployment'].replace({'A71':1, 'A72':2, 'A73':3, 'A74':4, 'A75':5}, inplace=True)
```

# Converting the binary nominal variable to numeric

```python
# Treating the binary nominal variable
DataForML['foreignWorker'].replace({'A201':1, 'A202':0}, inplace=True)

# Looking at data after nominal treatment
DataForML.head()
```

# Converting to get dummies

```python
# Treating all the nominal variables at once using dummy variables
DataForML_Numeric=pd.get_dummies(DataForML)

# Adding Target Variable to the data
DataForML_Numeric['creditRisks']=data['creditRisks']

# Printing sample rows
DataForML_Numeric.head()
```

# Split the data into tTraining and Testing sample 

```python
# Printing all the column names for our references
DataForML_Numeric.columns
```

```python
# Separate Target Variable and Predictor Variables
TargetVariable='creditRisks'
Predictors=['presentEmployment', 'foreignWorker', 'age', 'creditAmount',
       'durationMonth', 'checkingAccount_A11', 'checkingAccount_A12',
       'checkingAccount_A13', 'checkingAccount_A14', 'creditHistory_A30',
       'creditHistory_A31', 'creditHistory_A32', 'creditHistory_A33',
       'creditHistory_A34', 'purpose_A40', 'purpose_A41', 'purpose_A410',
       'purpose_A42', 'purpose_A43', 'purpose_A44', 'purpose_A45',
       'purpose_A46', 'purpose_A48', 'purpose_A49', 'savingsAccount_A61',
       'savingsAccount_A62', 'savingsAccount_A63', 'savingsAccount_A64',
       'savingsAccount_A65', 'personalStatus_A91', 'personalStatus_A92',
       'personalStatus_A93', 'personalStatus_A94', 'otherDebtors_A101',
       'otherDebtors_A102', 'otherDebtors_A103', 'property_A121',
       'property_A122', 'property_A123', 'property_A124',
       'otherInstallments_A141', 'otherInstallments_A142',
       'otherInstallments_A143', 'housing_A151', 'housing_A152',
       'housing_A153']
```

```python
X=DataForML_Numeric[Predictors].values
y=DataForML_Numeric[TargetVariable].values
```

```python
# split the data into training and testing set
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
scaledX_train = sc.fit_transform(X_train)
scaledX_test = sc.transform(X_test)
knn.fit(scaledX_train,y_train)
prediction = knn.predict(scaledX_test)
```

```python
print("What is the Training Accuracy")
print(knn.score(scaledX_train,y_train))

print("What is the Testing Accuracy")
print(knn.score(scaledX_test,y_test))
```
